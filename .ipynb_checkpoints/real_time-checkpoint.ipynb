```python
#!/usr/bin/env python

import pandas as pd
import numpy as np
import os
from takktile_ros.msg import Touch
import rospy
from std_msgs.msg import Float32
import time
import glob
import rospy
import torch
import torch.nn.functional as F
m = "/home/togzhan/sample/data/data/adaboost/biclass_adaboost2_max_depth150_n_estimators"
model = torch.load(m)
C1 = Touch()
C2 = Touch()
C3 = Touch()
C4 = Touch()
C5 = Touch()
C6 = Touch()
message = Touch()
#######################################################
def subscribe_contact(data):
    global C1, C2, C3, C4, C5
    C1 = data.pressure[21]
    C2 = data.pressure[27]
    C3 = data.pressure[15]
    C4 = data.pressure[9]
    C5 = data.pressure[4]
#######################################################
def listener():
    rospy.Subscriber("takktile/calibrated", Touch, subscribe_contact)
#######################################################

def rosbag_record(d):
#     d = "/home/togzhan/sample/real_time_data/"
    os.chdir(d)
    os.system("rosbag record /pressure_sub --duration=5")
#     time.sleep(6)
    dirs = glob.glob("*.bag")
    dirs.sort(key=lambda f: int(filter(str.isdigit, f)))
    l = len(dirs)
    str1 = 'rostopic echo /pressure_sub -b '
    str2 = dirs[l-1]
    print(dirs[l-1])
    print(dirs)
    str5 = os.path.splitext(dirs[l-1])[0]
    str3 = ' -p > ' + str5 + '.csv'
    str4 = str1+str2+str3
    
    os.system(str4)
    
#######################################################
#######################################################
def normalize(data):
    m = data.max()
    data = pd.DataFrame(data)
#     df_max = [420, 420, 420, 418, 418, 450, 432, 432, 432, 390, 472, 480, 480, 480, 435, 433, 501, 405, 400, 400, 400 ,400, 370, 455 ,444, 310, 445]
    df_max = [450]
    df_max = pd.DataFrame(df_max)
    df_min = 0;
    p = pd.DataFrame()
    k = []
    for i in range(27):
        k = np.divide(data.values[:,i].astype('float'), df_max.values[0].astype('float'))
        k = pd.DataFrame(k)    
        p = pd.concat([p, k], axis = 1)
        
    data = p
    data = pd.DataFrame(p)
    return data
#######################################################
######################################################
def real_time(d):
    rosbag_record(d)
    d = "/home/togzhan/sample/real_time_data/"
    directory = os.path.join(d)
    os.chdir(d)
    os.walk(directory)
    dirs = glob.glob("*.csv")
    l = len(dirs)
    # print(dirs)
    pressure1 = pd.DataFrame()
    pressure2 = pd.DataFrame()
    accelerometer = pd.DataFrame()

    dirs.sort(key=lambda f: int(filter(str.isdigit, f)))
    # listToStr = ' '.join(map(str, dirs[l-1]))
    diR = directory + dirs[l-1] 
    
    data = pd.read_csv(diR, sep=',', header = None, index_col=None)
    data_a = np.array(data)
    ############
    pressure1 = data_a[1:745, 4:23]
    pressure2 = data_a[1:745, 26:34]
    pressure = np.concatenate((pressure1, pressure2), axis = 1)
    pressure_n = normalize(pressure)
    
    P = []
    for i in range(27):
        P = np.concatenate([P[:], pressure_n.values[:,i]])
    ############
    accelerometer = data_a[1:745, 82:282]  
    A = []
    for i in range(200):
        A = np.concatenate([A[:], accelerometer[:,i]])
    ############
    imu = data_a[1:745, 34:52] 
    I = []
    for i in range(18):
        I = np.concatenate([I[:], imu[:,i]])
    data_all = np.concatenate([P[:],A[:],I[:]])
    data_all = pd.DataFrame(data_all)

    print data_all.shape
    print data_all.shape
    return data_all
    
def talker():
    while not rospy.is_shutdown():
        
        if(C2 > 200):
            print "here"
#             K = np.array(C)
            clas = pd.DataFrame()
# #             clas = real_time(d)
#             train_x	= torch.FloatTensor(clas.values[:].astype('float64'))
#             print model.predict(clas.transpose())
            
            #classify
            pub.publish(model.predict(clas.transpose()))
            time.sleep(1)
            print (C1<50) & (C2<50)& (C3<50)& (C4<50)& (C5<50)
            wU = True
            while wU == True:
                if (C1<50) & (C2<50)& (C3<50)& (C4<50)& (C5<50): 
#                     print "released"
                    pub2.publish(1)
                    wU = False
                else:
#                     print "waiting to release"
                    pub2.publish(0)       
                
        rate.sleep()     
if __name__ == '__main__':
    d = "/home/togzhan/sample/real_time_data/"
    
    rospy.init_node('glove', anonymous=True)
    pub = rospy.Publisher('class', Float32, queue_size =100) #pressure_sub
    pub2 = rospy.Publisher('release', Float32, queue_size =100) #pressure_sub
    rate = rospy.Rate(10) # 10hz
    listener()
    try:
        talker()
    except rospy.ROSInterruptException:
        rospy.spin()

        
```
